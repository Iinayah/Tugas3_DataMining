```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("penjualan.csv")
data.head()
```

```python

data.info()
```

Bar Chart dan Pie Chart Jumlah Produksi berdasarkan Kabupaten 

```python
unique_nama_kabupaten_kota = data['nama_kabupaten_kota'].unique()
unique_nama_kabupaten_kota.sort()
unique_nama_kabupaten_kota
```

```python
jml_nama_kabupaten_kota = []
for i in unique_nama_kabupaten_kota:
  jml_nama_kabupaten_kota.append(data[data['nama_kabupaten_kota']==i].jumlah_produksi.sum())
```

```python
plt.figure(figsize=(7,7))
plt.barh(unique_nama_kabupaten_kota, jml_nama_kabupaten_kota)
plt.title('Jumlah produksi mangga berdasarkan kabupaten kota')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(jml_nama_kabupaten_kota)
plt.legend(unique_nama_kabupaten_kota)
plt.title('Jumlah produksi mangga berdasarkan kabupaten kota')
plt.show()
```
